# Problem 3. Computing private information problems

## A. Private information

We seek to solve an essentially functional problem:

\begin{eqnarray}
    \max_{\pi}\sum_{A,Q,C} \pi(a, q, c)\cdot\left(q-c\right)\\
    \nonumber
    \text{subject to }\\
    \label{eqn:probsimplex}
    \sum_{A,Q,C}\pi(a,q,c)  & = &  1\\
    \label{eqn:exog}
    \forall\overline{a},\overline{q}\in A\times Q:\text{  } \frac{\sum_C\pi(\overline{a}, \overline{q}, c)}{\sum_{Q,C}\pi(\overline{a}, q, c)} & = & p(\overline{q}|\overline{a})\\
    \label{eqn:promise}
    \sum_{A,Q,C}\pi(a,q,c)U(a,c) & \geq & w_0\\
    \label{eqn:truth}
    \forall a, \hat{a}\in A:\text{  } \sum_{Q,C}\pi(a,q,c)U(a,c) & \geq & \sum_{Q,C}\pi(a,q,c)\frac{p(q|\hat{a})}{p(q|a)}U(\hat{a},c).
\end{eqnarray}

For computational purposes, we assume that $A = \{.2, .4\}$, $Q = \{1, 2\}$, and we discretize consumption in an evenly space grid of size twenty between $.1$ and $2$. Conditional probabilities are

\begin{eqnarray*}
    p(q = 1| a = .2) & = & .8\\
    p(q = 2| a = .2) & = & .2\\
    p(q = 1| a = .4) & = & .2\\
    p(q = 2| a = .4) & = & .8.
\end{eqnarray*}

Finally, utility function to be considered here is

\begin{equation}
    U(a,c) = \sqrt{c}(1-a).
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import linprog
```

### 1. Setting up the problem

This portion of the code initiates the parameters and utility functions for the problem.

```python
def generate_grid():
    a = np.array([.2, .4])
    qstates = np.array([1, 2])
    cgrid = np.linspace(.1, 2, 20)
    w0grid = np.linspace(0, 1.1, 20)
    
    return a, qstates, cgrid, w0grid, len(a), len(qstates), len(cgrid), len(w0grid)

def generate_params():
    """
        Spits out a vector [p(q=1|a=.2), p(q=2|a=.2), p(q=1|a=.4), p(q=2|a=.4)], 
    so precedence order of states is (A, Q, C)
    """
    return np.array([.8, .2, .2, .8])

def ufunc(a, c):
    """
        Inserts two arrays, a and c. Possibly different dimensions. 
    Returns a matrix with shape (len(a), len(c)) which is the utility 
    function at each value of a and c.
    """
    aux = a.ravel()
    caux = c.ravel()
    if (caux < 0).sum() > 0:
        raise ValueError('Consumption cannot be negative!')
    return np.kron((1-aux), np.sqrt(caux)).reshape(len(aux), len(caux))
```

Function linprog from scipy.optimize is an algorithm to solve the following kind of linear programming problem

\begin{eqnarray}
    \label{eqn:linprog}
    \min_{x\in\mathbb{R^n}} f^tx\\
    \nonumber
    \text{subject to}\\
    A_{ub}x & \leq & b_{ub} \\
    A_{eq}x & = & b_{eq} \\
    l \leq & x & \leq u.
\end{eqnarray}

Equivalently, any $x$ solving program \eqref{eqn:linprog} also solves
\begin{eqnarray}
    \label{eqn:linprog2}
    \max_{x\in\mathbb{R^n}} -f^tx\\
    \nonumber
    \text{subject to}\\
    A_{ub}x & \leq & b_{ub} \\
    A_{eq}x & = & b_{eq} \\
    l \leq & x & \leq u.
\end{eqnarray}

Programmatically, we vectorize the 3-dimensional tensor $\pi(a_i,q_j,c_k) = \pi_{ijk}$ as follows:

\begin{equation}
    \pi_{vec} =
    \begin{bmatrix}
        \pi_{0,0,0}\\
        \pi_{0,0,1}\\
        \vdots\\
        \pi_{0,0,19}\\
        \pi_{0,1,0}\\
        \vdots\\
        \pi_{0,1,19}\\
        \vdots\\
        \pi_{1,1,19}
    \end{bmatrix}
\end{equation}

This is a vector in $\mathbb{R}^{2\times 2\times 20}=\mathbb{R}^{80}$. For that end, we use the numpy.ravel function. Similarly the matrix $f^t$ consists in all $2\times 20$ possible combinations of $c-q$, but repeated for each value of $a$ possible. Formally, the vector $f$ is the result of stacking the vector

\begin{equation}
    v = 
    \begin{bmatrix}
        -q_0+c_0\\
        -q_0+c_1\\
        -q_0+c_2\\
        \vdots\\
        -q_0+c_{19}\\
        -q_1+c_0\\
        \vdots\\
        -q_1+c_{19}
    \end{bmatrix}
\end{equation}
len(a) times. Formally

\begin{equation}
    \label{eqn:fvec}
    f = np.ones(len(a)) \otimes v.
\end{equation}

That deals with the objective function. Now let us deal with the inequality constraints, expressed in inequalities \eqref{eqn:promise} and \eqref{eqn:truth} in the introductory part of this section. Inequality \eqref{eqn:promise} is one-dimensional for each w0 in the w0 grid. For starts, we need to ravel u(a,c) to get it in shape len(a) * len(c). Then we do the Kronecker operation, just like we did in \eqref{eqn:fvec}. Set of inequalities \eqref{eqn:truth} are 2 in number. It can be written as

\begin{equation}
    \label{eqn:promtruth}
    \begin{bmatrix}
        u_{0,0} - \frac{p_{1,0}}{p_{0,0}}u_{1,1} & 0\\
        u_{0,1} - \frac{p_{1,0}}{p_{0,0}}u_{1,2} & 0\\
        u_{0,2} - \frac{p_{1,0}}{p_{0,0}}u_{1,3} & 0\\
        u_{0,3} - \frac{p_{1,0}}{p_{0,0}}u_{1,4} & 0\\
        \vdots & \vdots\\
        u_{0,18} - \frac{p_{1,1}}{p_{0,1}}u_{1,18} & 0\\
        u_{0,19} - \frac{p_{1,1}}{p_{0,1}}u_{1,19} & 0\\
        0 & u_{1,0} - \frac{p_{0,0}}{p_{1,0}}u_{0,0}\\
        \vdots & \vdots\\
        0 & u_{1,19} - \frac{p_{0,1}}{p_{1,1}}u_{0,19}
    \end{bmatrix}^t
    \pi_{vec}
    \geq
    0_{2\times 1}
\end{equation}

To construct the matrix $A_{ub}$ we write the first line as the negative of $u(a,c).ravel()$. In the end, the matrix $A_{ub}$ must be

\begin{equation}
    \label{eqn:Aub}
    \begin{bmatrix}
        -u_{0,0} & -u_{0,0} + \frac{p_{1,0}}{p_{0,0}}u_{1,0} & 0\\
        -u_{0,1} & -u_{0,1} + \frac{p_{1,0}}{p_{0,0}}u_{1,1} & 0\\
        -u_{0,2} & -u_{0,2} + \frac{p_{1,0}}{p_{0,0}}u_{1,2} & 0\\
        -u_{0,3} & -u_{0,3} + \frac{p_{1,0}}{p_{0,0}}u_{1,3} & 0\\
        \vdots &\vdots & \vdots\\
        -u_{0,18} & -u_{0,18} + \frac{p_{1,1}}{p_{0,1}}u_{1,18} & 0\\
        -u_{0,19} & -u_{0,19} + \frac{p_{1,1}}{p_{0,1}}u_{1,19} & 0\\
        -u_{1,0} & 0 & -u_{1,0} + \frac{p_{0,0}}{p_{1,0}}u_{0,0}\\
        \vdots & \vdots & \vdots\\
        -u_{1,19} & 0 & -u_{1,19} + \frac{p_{0,1}}{p_{1,1}}u_{0,19}
    \end{bmatrix}^t
\end{equation}

and $b_{ub}$ has to be

\begin{equation}
    \label{eqn:bub}
    b_{ub} = 
    \begin{bmatrix}
        -w_0\\
        0\\
        0
    \end{bmatrix}
\end{equation}

The first line of $A_{ub}$ is easy. Just ravel $u(a,c)$ and then stack it len(q) times. The second and third lines can be obtained by, first, computing $u(0,cgrid)$ and $u(1,cgrid)$, then do the kronecker products of each of those with np.ones(len(q)) from the right: $u \otimes np.ones(len(q))$. Next, take the vectors $[p_{1,0}, p_{1,1}]$ and $[p_{0,0}, p_{0,1}]$ and divide each of those by $[p_{0,0}, p_{0,1}]$ and $[p_{1,0}, p_{1,1}]$. At last, do kronecker product of each of those by np.ones(len(c)) from the left.

To complete, we need a set of len(q) * len(a) + 1 equations, given by \eqref{eqn:probsimplex} and \eqref{eqn:exog} in the beginning of this section. Equation \eqref{eqn:exog} can be rewritten as the following system:

\begin{equation}
    \begin{bmatrix}
        1 & 1 & \cdots & 1 & 0 & \cdots & 0\\
        0 & 0 & \cdots & 0 & 1 & \cdots & 0\\
        \vdots & \vdots & \ddots & \vdots & \vdots & \ddots & \vdots\\
        0 & 0 & \cdots & 0 & 0 & \cdots & 1
    \end{bmatrix}
    \pi_{vec} = 
    \begin{bmatrix}
        p_{0,0} & p_{0,0} & \cdots & p_{0,0} & \cdots & 0\\
        \vdots & \vdots & \ddots & \vdots & \ddots &\cdots\\
        0 & 0 & \cdots & 0 & \cdots & p_{1,1}
    \end{bmatrix}
    \pi_{vec}
\end{equation}

in which the matrix on the left has len(c) consecutive 1's repeated in each row, but shifting c by c, whereas the matrix on the right has repeated $p_{i,j}$ len(q) * len(c) times. Notice, however, that for each $a$, one of the equations in \eqref{eqn:exog} is redundant. Using the fact that $nq=2$ to get a better picture, we should just care about

\begin{equation}
    \begin{bmatrix}
        1 - p_{0,0} & 1 - p_{0,0} & 1 - p_{0,0} & 1 - p_{0,0} & 1 - p_{0,0} & \cdots & 0 & 0 & 0\\
        0 & 0 & 0 & 0 & 0 & \cdots & 1 - p_{1,1} & 1 - p_{1,1} & 1 - p_{1,1}
    \end{bmatrix}
    \pi_{vec} = 
    \begin{bmatrix}
        0\\
        0
    \end{bmatrix}
\end{equation}

in which, in first row, $1-p_{0,0}$ is repeated len(c) times, $-p_{0,0}$ is repeated len(q) * len(c) - len(c) times, and then 0 is repeated the other len(a) * len(q) * len(c) - len(q) * len(c) times. In the second row, we have len(a) * len(q) * len(c) - len(q) * len(c) repetitions of 0, then len(q) * len(c) - len(c) repetitions of $-p_{1,1}$, then the rest is just repeated $1 - p_{1,1}$.

```python
def solve(info = False, method = 'interior-point'):
    """
        This function relies on na = nq = 2. That's bad. 
    Will correct it later.
    """
    a, qstates, cgrid, w0grid, na, nq, nc, n0 = generate_grid()
    
    cvec = np.kron(np.ones(nq), cgrid)
    qvec = np.kron(qstates, np.ones(nc))
    v = cvec - qvec
    
    #### Matrix f
    f = np.kron(np.ones(na), v)
    
    #### Generating matrix Aub. 
    u = ufunc(a,cgrid)
    u = np.kron(np.ones(nq), u)
    
    p = generate_params()
    aub1 = np.zeros(na * nq * nc)
    aub1[:nq * nc] = -u[0,:]
    aub1[-nq * nc:] = -u[1,:]
    if info == False:
        p0 = np.array([p[2]/p[0], p[3]/p[1]])
        p1 = 1/p0

        p01 = np.kron(p0, np.ones(nc))
        p10 = np.kron(p1, np.ones(nc))
        
        aub2 = np.zeros(na * nq * nc)
        aub3 = np.zeros(na * nq * nc)
        
        aub2[:nq * nc] = p01 * u[1,:] - u[0,:]
        aub3[-nq * nc:] = p10 * u[0,:] - u[1,:]

        #### Matrix A_{ub}
        aub = np.array([aub1, aub2, aub3])
        bub = np.zeros(1 + na)
    else:
        aub = aub1.reshape(1, na * nq * nc)
        bub = 0
    
    #### Now, construct matrix A_eq
    fr = np.append(np.ones(nc), np.zeros(na * nq * nc - nc))
    sr = np.append(np.zeros(na * nq * nc - nc), np.ones(nc))
    ls = np.vstack([fr, sr])
    pvals = np.array([p[0], p[-1]])
    pvals = np.kron(pvals, np.ones(na * nq * nc)).reshape(2, na * nq * nc)
    auxmtx = np.kron(np.identity(na), np.ones(nq * nc))
    rs = auxmtx * pvals
    
    aeq = np.vstack([np.ones(na * nq * nc), ls - rs])
    beq = np.zeros(na + 1)
    beq[0] = 1
    
    result = np.ndarray(n0, dtype='object')
    pi = np.zeros((na * nq * nc, n0))
    for iw, w0 in enumerate(w0grid):
        if info == False:
            bub[0] = -w0
        else:
            bub = -w0
        
        result[iw] = linprog(c = f, A_ub = aub, b_ub = bub, A_eq = aeq, b_eq = beq, bounds = (0, 1), method = method)
        pi[:,iw] = result[iw].x
        
    
    return pi.reshape((na, nq, nc, n0)), result
```

```python
pi, result = solve(method = 'simplex')
```

### 2. Results

The variable result is an array of objects from the class scipy.linprog. For each iw, w0 in enumerate(w0grid) result[iw] is an instance of this class for the planner's problem with outside option w0grid[iw] = w0. We are particularly interested in the argmax, represented as the 4-dimensional tensor pi and in the slackness of the constraints, summarized by the attribute slack from the class scipy.linprog. The first element of result.slack is the slackness of the participation constaint, the second one is the slackness of low-effort incentive compatibility. Finally, the third one should be the slackness of the high-effort IC. The closer this parameter is to zero, the closer the constraint is to being active (binding).

Now we proceed to show all results in the private information problem.

This first part plots the slackness variables as a function of the reservation utility $w_0$

```python
a, qstates, cgrid, w0grid, na, nq, nc, n0 = generate_grid()

slacks = np.zeros((na + 1, n0))
for iw in range(len(w0grid)):
    slacks[:, iw] = result[iw].slack


fig, ax = plt.subplots()
ax.plot(w0grid, slacks.T)
ax.set_title('Private information - Slackness', fontsize=14)
ax.set_xlabel('Outside option ($w_0$)', fontsize=14)
ax.set_ylabel('Slackness', fontsize=14)
ax.legend(labels=['IR', 'IC Low', 'IC High'])
plt.show()
```

As can be seen, individual rationality (IR) is slack for low values of $w_0$ and, as the outside option increases, the slackness diminishes (blue line). The incentive compatibility constraint for high effort is always binding (green line) since high effort is costly (moral hazard). As for the incentive compatibility for low types, it is binding for low outside options. It seems that when the planner suggests low effort to someone when outside option is low, the agent has some incentive to put more effort in equilibrium.

Next, we show the different levels of consumption for each $w_0$, conditional on effort $a$ and output $q$. By inspection of results below (the $x$'s corresponding to vectorized $\pi$'s), it seems that consumption is not being randomized. In fact, results clearly show large proportions of weights equal to zero and big concentration of positive entries in vectorized $\pi$.

```python
for iw in range(n0):
    print(result[iw])
```

I also would find instructive to plot the average effort for each $w_0$.

```python
fig, ax = plt.subplots()

pivec = pi.reshape(na * nq * nc, n0).T
avec = np.kron(a, np.ones(nq * nc))

effs = pivec @ avec

hlinelow = a[0] * np.ones(n0)
hlinehigh = a[1] * np.ones(n0)

ax.plot(w0grid, effs, lw=2)
ax.plot(w0grid, np.array([hlinelow, hlinehigh]).T, 'k--', lw=1)
ax.set_title('Private Information - Average effort', fontsize=14)
ax.set_xlabel('Outside option ($w_0$)', fontsize=14)
ax.set_ylabel('Effort ($a$)', fontsize=14)
plt.show()
```

## B. Full information

Below we solve the full information problem. Here we dispose of the incentive compatibility constraints, since now the planner can impose whichever level of effort he wants.

```python
pifull, resultfull = solve(info = True, method = 'simplex')
```

```python
slacksfull = np.zeros(n0)
for iw in range(len(w0grid)):
    slacksfull[iw] = resultfull[iw].slack


fig, ax = plt.subplots()
ax.plot(w0grid, slacksfull)
ax.set_title('Full information - slackness', fontsize=14)
ax.set_xlabel('outside option ($w_0$)', fontsize=14)
ax.set_ylabel('slackness (IR)', fontsize=14)
plt.show()
```

Not surprisingly, the IR slackness decreases as the outside option increases.

Below we analyze the results with full information. Again, consumption does not seem to be randomized.

```python
for iw in range(n0):
    print(resultfull[iw])
```

Again, it is useful to plot average effort.

```python
fig, ax = plt.subplots()

pifvec = pifull.reshape(na * nq * nc, n0).T

efffs = pifvec @ avec

hlinelow = a[0] * np.ones(n0)
hlinehigh = a[1] * np.ones(n0)

ax.plot(w0grid, efffs, lw=2)
ax.plot(w0grid, np.array([hlinelow, hlinehigh]).T, 'k--', lw=1)
ax.set_title('Full Information - Average effort', fontsize=14)
ax.set_xlabel('Outside option ($w_0$)', fontsize=14)
ax.set_ylabel('Effort ($a$)', fontsize=14)
plt.show()
```

```python
qvec = np.kron(np.kron(np.ones(na), qstates), np.ones(nc))
cvec = np.kron(np.ones(na * nq), cgrid)

ufvec = pifvec @ (qvec - cvec)
uvec = pivec @ (qvec - cvec)

fig, ax = plt.subplots()
ax.plot(w0grid, uvec, label = 'Private')
ax.plot(w0grid, ufvec, label = 'Full')
ax.legend()
plt.show()

plt.plot(w0grid, ufvec - uvec)
plt.show()
```

## C. Graphs for discussion

To compare private and full information scenarios, we choose to plot both average efforts and average consumptions.

```python
cvec = np.kron(np.ones(na * nq), cgrid)

cees = pivec @ cvec
ceesf = pifvec @ cvec

hlineclow = cgrid[0] * np.ones(n0)
hlinechigh = cgrid[19] * np.ones(n0)

fig, ax = plt.subplots()

ax.plot(w0grid, np.array([effs, efffs]).T, lw=2)
ax.plot(w0grid, np.array([hlinelow, hlinehigh]).T, 'k--', lw=1)
ax.set_title('Average effort', fontsize=14)
ax.set_xlabel('Outside option ($w_0$)', fontsize=14)
ax.set_ylabel('Effort ($a$)', fontsize=14)
ax.legend(labels = ['private info', 'full info'])
plt.show()

fig, ax = plt.subplots()

ax.plot(w0grid, np.array([cees, ceesf]).T, lw=2)
ax.plot(w0grid, np.array([hlineclow, hlinechigh]).T, 'k--', lw=1)
ax.set_title('Average consumption', fontsize=14)
ax.set_xlabel('Outside option ($w_0$)', fontsize=14)
ax.set_ylabel('Consumption ($c$)', fontsize=14)
ax.legend(labels = ['private info', 'full info'])
plt.show()
```
